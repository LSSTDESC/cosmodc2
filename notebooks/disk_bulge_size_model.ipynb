```python
from cosmodc2.size_modeling import median_size_vs_luminosity_late_type
from cosmodc2.size_modeling import mc_size_vs_luminosity_early_type
from cosmodc2.size_modeling import median_size_vs_luminosity_early_type
from cosmodc2.size_modeling import mc_size_vs_luminosity_late_type

magr_table = np.linspace(-25, -10, 500)

median_disk_size_z0 = median_size_vs_luminosity_late_type(magr_table, 0)
median_disk_size_z0p5 = median_size_vs_luminosity_late_type(magr_table, 0.5)
median_disk_size_z1 = median_size_vs_luminosity_late_type(magr_table, 1)
median_disk_size_z2 = median_size_vs_luminosity_late_type(magr_table, 2)

median_bulge_size_z0 = median_size_vs_luminosity_early_type(magr_table, 0)
median_bulge_size_z0p5 = median_size_vs_luminosity_early_type(magr_table, 0.5)
median_bulge_size_z1 = median_size_vs_luminosity_early_type(magr_table, 1)
median_bulge_size_z2 = median_size_vs_luminosity_early_type(magr_table, 2)
```

```python
ngals = int(2e4)
magr_mock = np.random.uniform(-25, -15, ngals)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0)  #  smash top and bottom plots
fig.subplots_adjust(wspace=0)  #  smash left and right plots

yscale1 = ax1.set_yscale('log')
yscale2 = ax2.set_yscale('log')

__=ax1.scatter(magr_mock, mc_size_vs_luminosity_early_type(magr_mock, 0), s=0.1, color='red')
__=ax1.plot(magr_table, median_bulge_size_z0, color='orange', lw=2, 
            label=r'${\rm median\ z=0\ bulge\ size}$')
__=ax1.plot(magr_table, median_disk_size_z0, color='cyan', lw=2, 
           label=r'${\rm median\ z=0\ disk\ size}$')

__=ax2.scatter(magr_mock, mc_size_vs_luminosity_late_type(magr_mock, 0), s=0.1, color='blue')
__=ax2.plot(magr_table, median_disk_size_z0, color='cyan', lw=2, 
           label=r'${\rm median\ z=0\ disk\ size}$')
__=ax2.plot(magr_table, median_bulge_size_z0, color='orange', lw=2, 
            label=r'${\rm median\ z=0\ bulge\ size}$')

xlim1 = ax1.set_xlim(-15, -25)
xlim2 = ax2.set_xlim(-15, -25)

legend1 = ax1.legend()
legend2 = ax2.legend()
xlabel1 = ax1.set_xlabel(r'$M_{\rm r}$')
xlabel2 = ax2.set_xlabel(r'$M_{\rm r}$')
ylabel1 = ax1.set_ylabel(r'${\rm R_{50}\ [kpc]}$')

figname = 'z0_bulge_disk_sizes.pdf'
fig.savefig(figname, bbox_extra_artists=[xlabel1, ylabel1], bbox_inches='tight')
```
